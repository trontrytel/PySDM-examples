```python
from PySDM_examples.utils.show_plot import show_plot
from PySDM_examples.Alpert_and_Knopf_2016 import Isothermal, Table1
from PySDM.physics import si
```

```python
cases = Table1()
```

### Fig 2 a

```python
sim_2a = Isothermal(cases=cases, dt=10*si.s, total_time=14*si.min)
sim_2a.run(('IsoBR',))
```

```python
sim_2a.plot(ylim=(.5, 1))
show_plot()
```

### Fig 2 b

```python
sim_2b = Isothermal(cases=cases, dt=10*si.s, total_time=18*si.min)
sim_2b.run(('IsoHE1',))
```

```python
sim_2b.plot(ylim=(1e-2, 1))
show_plot()
```

### Fig 2 c

```python
sim_2c = Isothermal(cases=cases, dt=100*si.s, total_time=150*si.min)
sim_2c.run(('IsoHE2',))
```

```python
sim_2c.plot(ylim=(1e-2, 1))
show_plot()
```
